```python

import os

os.environ["KERAS_BACKEND"] = "tensorflow"

import keras
import tensorflow as tf
import numpy as np
from keras import layers

import string
import re
import csv

#Loads model with the highest accuracy
model = tf.keras.models.load_model('../model/ahole-model_unbalanced_3-epochs.keras')
#Display model summary for each layer
print(model.summary())
#Find and safe model weights
weights = []
for lay in model.layers: #for each layer:
    print(lay.name)
    weights.append([lay.name]) #add name to output file
    for i in range(len(lay.get_weights())): #in each array of weights
        print(len(lay.get_weights()[i])) #print how many NumPy arrays
        weights.append([(lay.get_weights()[i])]) #add to weight output
```

```python
#Adds each row of output to model-weights.csv
if os.path.exists('../model-weights.csv'):
    with open('../model-weights.csv', 'w', newline='') as csvfile:
            writer = csv.writer(csvfile)
            for row in weights:
                    writer.writerow(row)
```
