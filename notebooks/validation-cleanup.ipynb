```python
import pandas as pd
import os
import re
```

```python
stat_blocks = [f'../validation-stats/{file}' for file in os.listdir('../validation-stats') if 'DS_Store' not in file]
```

```python
stats = pd.concat([pd.read_csv(file) for file in stat_blocks]).reset_index(drop=True)
```

```python
stats['epochs'] = stats['model'].apply(lambda file: re.search('([0-9]+)-epochs', file).group(1))
stats['model_type'] = stats['model'].apply(lambda file: file.split('_')[0].split('/')[-1])
stats['balance'] = stats['model'].apply(lambda file: file.split('_')[1])
stats['model_type'] = stats['model_type'].replace('ahole-model', 'neural model')
```

```python
stats = stats[['model_type', 'balance', 'epochs', 'accuracy_rate', 'model']].rename({
    'model': 'filepath'
}, axis=1)
```

```python
stats = stats.sort_values('accuracy_rate', ascending=False).reset_index(drop=True)
```

```python
stats['accuracy_rate'] = stats['accuracy_rate'].apply(lambda x: round(x * 100, 1)).apply(lambda y: f'{y}%')
```

```python
stats.to_csv('../validation-stats.csv', index=False)
```
